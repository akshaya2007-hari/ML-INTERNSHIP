```python
import numpy as np
import tensorflow as tf
import keras
from keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from keras.utils import to_categorical
```

```python
sentences=[
    "I love my dog",
    "I love my cat",
    "You love my dog",
    "I like to learn ML",
    "I enjoy learning new things",
    "At can perform many task",
    "Artificial Intelligence is the future",
    "Machine Learning is the future",
    "Deep Learning is the future"


]
```

```python
my_tokenizer=Tokenizer()
my_tokenizer.fit_on_texts(sentences)
```

```python
my_tokenizer.word_index
```

```python
total_words=len(my_tokenizer.word_index)+1
total_words
```

```python
input_sequences=[]
for i in sentences:
  token_list=my_tokenizer.texts_to_sequences([i])[0]
  for j in range(1,len(token_list)):
    n_gram_sequence=token_list[:j+1]
    input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
max_sequence_len=max([len(x) for x in input_sequences])
print(max_sequence_len)
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
```

```python
input_sequences
```

```python
X_train=input_sequences[:,:-1]
label=input_sequences[:,-1]
```

```python
print(X_train[2])
print(label[2])
```

```python
output_c=to_categorical(label,num_classes=total_words)
output_c[2]
```

```python
RNN=keras.Sequential([
    layers.Embedding(total_words,100,input_length=max_sequence_len-1),
    layers.SimpleRNN(150),
    layers.Dense(total_words,activation='softmax')

])
```

```python
RNN.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
RNN.fit(X_train,output_c,epochs=100,verbose=1)
```

```python
seed_text = __builtins__.input("Enter the text: ")
token_list = my_tokenizer.texts_to_sequences([seed_text])[0]
token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
predicted = RNN.predict(token_list)
predicted
```

```python
predicted=np.argmax(predicted,axis=1)
print(predicted)
```

```python
predicted_word=""
for word,index in my_tokenizer.word_index.items():
  if index==predicted:
    predicted_word=word
    break
print(f"The next word is:{predicted_word}")
```
