```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    'Maths': [80, 90, 85, 70],
    'Physics': [75, 88, 82, 68],
    'Chemistry': [78, 85, 80, 65]
}

df = pd.DataFrame(data)

# Correlation matrix
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Subject Marks Correlation Heatmap")
plt.show()
```
